# seperation of dependent and independent variables

```python
x=df.drop(columns=["Attrition"],axis =1)
y=df.iloc[:,1:2]
```

```python
x.head()
```

```python
y.head()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Assuming x is a DataFrame containing multiple columns
columns_to_encode = [ "BusinessTravel","Department", "EducationField", "Gender", "JobRole","MaritalStatus","Over18","OverTime"]

for column in columns_to_encode:
    x[column] = le.fit_transform(x[column])
```

```python
x.head()
```

```python
y.head()
```

# splitting the dataset to train set and training set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# feature scaleing

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python


y_pred = classifier.predict(x_test)

# Convert y_test DataFrame to a NumPy array before reshaping
y_test_array = y_test.to_numpy()

# Now, you can reshape the arrays and concatenate them
result = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test_array.reshape(len(y_test_array), 1)), axis=1)

# Print the result
print(result)

```
